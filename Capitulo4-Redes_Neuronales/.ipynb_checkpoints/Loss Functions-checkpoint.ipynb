```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Mean Squared Error Loss Function:

```python
y = 0.5
```

```python
y_pred = np.arange(0,1.1,0.1)
y_pred
```

```python
plt.figure(figsize=(8,6))
for i in y_pred:
    MSE = (1/len(y_pred))*(i - y)**2
    plt.scatter(i, MSE, color='blue')
```

## Binary Cross Entropy Loss Function:

```python
plt.figure(figsize=(8,6))
for i in y_pred:
    if i == 0 or i == 1:
        BCE = 0
    else:    
        BCE = -(y * np.log(i) + (1-y) * np.log(1-i))
        plt.scatter(i, BCE, color='red')    
```

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
def cost_1(x):
    return -np.log(sigmoid(x))
```

```python
def cost_0(x):
    return -np.log(1-sigmoid(x))
```

```python
z = np.arange(-10, 10, 0.1)
```

```python
c1 = [cost_1(x) for x in z]
c0 = [cost_0(x) for x in z]
```

```python
ct = []
for i in range(len(c1)):
    ct.append(c0[i]+c1[i])
```

```python
plt.figure(figsize=(8,6))
plt.plot(sigmoid(z), c1, linestyle='--', color='red', linewidth=2.0, label='Cost(w) if y=1')
plt.plot(sigmoid(z), c0, linestyle='--', color='green',linewidth=2.0,label='Cost(w) if y=0')
plt.plot(sigmoid(z), ct, color='blue',linewidth=2.0, label='Total Cost(w)')
plt.ylim(0.0, 5)
plt.xlabel('$\phi$z')
plt.ylabel('Cost(w)')
plt.legend(loc='best')
```
