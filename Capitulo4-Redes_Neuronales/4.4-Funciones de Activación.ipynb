```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
z = np.linspace(-1,1,20)
print(z)
```

## Identity Function:

```python
def f(x):
    return x
```

```python
ifunc = [f(x) for x in z]
```

```python
plt.figure(figsize=(8,6))
plt.plot(z, ifunc, color='red',linewidth=2.0, label="Identity Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")    
```

## Step Function:

```python
# This function activates the neuron if the activation if the activation
# is above a certain value that is called threshold activity function
def sf(x):
    if x >= 0:
        output = 1
    else:
        output = 0
    return output    
```

```python
sf_ = [sf(x) for x in z]
```

```python
plt.figure(figsize=(6,6))
plt.plot(z, sf_, color='green',linewidth=2.0, label="Step Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")
```

## Sigmoid Function:

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
X_sig = np.linspace(-10,10,100)
```

```python
sigmoid_ = [sigmoid(x) for x in X_sig]
```

```python
plt.figure(figsize=(8,6))
plt.plot(X_sig, sigmoid_, color='blue',linewidth=2.0, label="Sigmoid Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")
```

## Hyperbolic Tangent:

```python
ht = [np.tanh(x) for x in z]
```

```python
plt.figure(figsize=(6,6))
plt.plot(z, ht, color='purple', linewidth=2.0, label="Hyberbolic Tangent Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")
```

## ReLU (Rectified Linear Unit)

```python
def relu(x):
    if x >= 0:
        output = x
    else:
        output = 0
    return output  
```

```python
relu_ = [relu(x) for x in z]
```

```python
plt.figure(figsize=(8,6))
plt.plot(z, relu_, color='black',linewidth=2.0, label="ReLU Function")
plt.xlabel("Z")
plt.ylabel("f(z)")
plt.legend(loc="best")
```
