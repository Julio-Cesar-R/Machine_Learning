# Heart Disease Prediction

### Data Source: https://www.openml.org/d/1498

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
%matplotlib inline
```

### Atributes
1. Systolic blood pressure (Sbp)
2. Cumulative tobacco consumption (kg)
3. Low density lipoprotein (LDL-cholesterol)
4. Adiposity
5. Family history of heart disease (Present/Absent)
6. Type-A behavior
7. Obesity
8. Current alcohol consumption
9. Age during onset of condition
10. CHD response (Coronary Heart Disease)

## Loading data

```python
df = pd.read_csv('https://www.openml.org/data/get_csv/1592290/phpgNaxZe')
df.head()
```

```python
columns = ['sbp', 'tobacco', 'ldl', 'adiposity', 'famhist', 'type', 'obesity', 'alcohol', 'age', 'chd']
```

```python
df.columns = columns
df.head()
```

```python
df.describe()
```

## Scaling data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
df['famhist'] = encoder.fit_transform(df['famhist'])
df['chd'] = encoder.fit_transform(df['chd'])
df.head()
```

```python
df.describe()
```

## Visualizing features:

```python
sns.catplot(data=df, orient="v", palette='husl', kind="box", aspect=2.0)
```

```python
sns.pairplot(df, kind="reg", palette='husl', height=2, aspect=1.5)
```

```python
df.corr()
```

## Defining Variables

```python
X=df[['sbp','tobacco','ldl','adiposity','age']]
y=df['chd']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state= 2019)
```

## Predictions with Support Vector Machine:

```python
from sklearn import svm
svm_clf = svm.SVC(C=0.025, kernel='linear')
```

```python
svm_clf.fit(X_train, y_train)
```

```python
y_pred_svm = svm_clf.predict(X_test)
y_pred_svm
```

### Support Vector Machine Metrics:

```python
print(f"Accuracy Score:{accuracy_score(y_test, y_pred_svm)}")
```

```python
print(f"Confusion Matrix\n:{confusion_matrix(y_test, y_pred_svm)}")
```

```python
print(f"Classification Report\n:{classification_report(y_test, y_pred_svm)}")
```

## Predictions with k-Nearest Neighbor (KNN):

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_clf.fit(X_train, y_train)
```

```python
y_pred_knn = knn_clf.predict(X_test)
```

### KNN metrics

```python
print(f"Accuracy Score:{accuracy_score(y_test, y_pred_knn)}")
```

```python
print(f"Confusion Matrix\n:{confusion_matrix(y_test, y_pred_knn)}")
```

```python
print(f"Classification Report\n:{classification_report(y_test, y_pred_knn)}")
```

## Predictions Multilayer Perceptron:

```python
from sklearn.neural_network import MLPClassifier
mlp_clf = MLPClassifier(max_iter=1000)
```

```python
parameters = {'solver': ['sgd','lbfgs'],
              'activation': ['logistic', 'relu'],
             'alpha': [0.01, 0.001, 0.0001, 0.00001],
             'hidden_layer_sizes':(15,15,5),
             'random_state': [2019]}
```

```python
from sklearn.metrics import make_scorer
```

```python
acc_scorer = make_scorer(accuracy_score)
grid = GridSearchCV(mlp_clf, parameters, cv=5, iid=True, scoring=acc_scorer)#iid True, return the average score across folds
grid = grid.fit(X_train, y_train)
```

```python
mlp_clf = grid.best_estimator_
```

```python
mlp_clf.fit(X_train, y_train)
```

```python
y_pred_mlp = mlp_clf.predict(X_test)
```

### Multilayer Perceptron Metrics

```python
print(f"Accuracy Score:{accuracy_score(y_test, y_pred_mlp)}")
```

```python
print(f"Confusion Matrix\n:{confusion_matrix(y_test, y_pred_mlp)}")
```

```python
print(f"Classification Report\n:{classification_report(y_test, y_pred_mlp)}")
```
