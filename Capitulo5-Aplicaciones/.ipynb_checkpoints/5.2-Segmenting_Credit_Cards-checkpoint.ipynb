# Segmenting Credit Cards

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import ShuffleSplit
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Loading the data:

```python
#https://www.kaggle.com/arjunbhasin2013/ccdata
file = '../data/CC GENERAL.csv'
df = pd.read_csv(file)
df.head()
```

```python
df.shape
```

```python
print(df.dtypes)
```

## Cleaning the data:

```python
df = df.drop(['CUST_ID'], axis=1)
```

```python
#Filling missing values with the most commonly occurring value in the column
#df_clean = df.apply(lambda x:x.fillna(x.value_counts().index[0]))
df_clean = df.apply(lambda x:x.fillna(x.mean()))
```

## Feature Scaling:

```python
values = df_clean.values
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(values)
df_ = pd.DataFrame(x_scaled,columns=df_clean.columns)
df_.head()
```

## Number of Clusters: Elbow Method

## Splitting the data:

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df_)
    Sum_of_squared_distances.append(km.inertia_)
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal K')
plt.show()
```

## Splitting the data:

```python
np.random.seed(2019)
ids = np.random.rand(len(df_)) < 0.8
train = df_[ids]
test = df_[~ids]
```

```python
X_train = np.array(train)
X_test = np.array(test)
```

## Training the data:

```python
# n_clusters: 10
kmeans = KMeans(n_clusters=10, random_state=2019).fit(X_train)
```

## Predictions:

```python
pred = kmeans.predict(X_test)
```

```python
test = test.assign(Predicted_Cluster = pred)
test.head()
```

```python
df_summary = test.groupby(by='Predicted_Cluster').mean()
df_summary = df_summary[['BALANCE', 'PURCHASES', 
                               'PURCHASES_FREQUENCY','CREDIT_LIMIT', 
                               'ONEOFF_PURCHASES_FREQUENCY', 
                              'MINIMUM_PAYMENTS','PRC_FULL_PAYMENT', 
                               'PAYMENTS']]
df_summary
```
