```python
#librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick#Personalizacion de graficos
%matplotlib inline
```

## Loading the data:

dataset: https://www.kaggle.com/shivachandel/kc-house-data

Ejecutar codigo con texto enriquecido :https://miktex.org/download

Colores de matplotlib :https://matplotlib.org/stable/tutorials/colors/colormaps.html#sphx-glr-tutorials-colors-colormaps-py

```python
#Cargamos el archivo
data = pd.read_csv('../data/kc_house_data.csv')
df = data.sample(100)
df.head()#Mostramos el archivo
```

```python
#Asignamos valores a los ejes
x = df['sqft_living']
y = df['price']
```

## Visualizing the data with default Matplotlib

```python
#Visualizamos el grafico
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(x, y, color='green')
```

## Improving visualization:

```python
# Remove Spines
#Remover los bordes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
fig
```

```python
# Using Latex : www.latex-project.org
#Labels enriquecidos
ax.set_title(r'House Size in $ft2$ vs. Price', fontsize=20, pad=20)
ax.set_xlabel('House Size', fontsize=16, labelpad=20)
ax.set_ylabel('Price', fontsize=16, labelpad=20)
fig
```

```python
#Add units to Axis labels
#Asignar en una variable el font del label "y_format"
y_format = '${x:,.2f}'
y_tick = mtick.StrMethodFormatter(y_format)
ax.yaxis.set_major_formatter(y_tick)

x_format = r'{x:,.0f} $ft^2$'
x_tick = mtick.StrMethodFormatter(x_format)
ax.xaxis.set_major_formatter(x_tick)
fig
```

```python
# Change color, marker, size and transparency
#Crear un marcador por cada punto
ax.scatter(x, y, s=200, marker='+', color='#FF5733', alpha=0.8)
fig
```

```python
# adding major gridlines
#Agregamos una cuadricula con "grid"
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
fig
```

```python
# applying color maps
#Agregar marcas a los puntos
ax.scatter(x, y, s=300, c=y, marker='+', cmap='Greens', alpha=0.8)
fig
```

```python
#Cambiar color de los puntos y los marcadores
ax.scatter(x, y, s=300, c=y, marker='+', cmap='Reds', alpha=0.8)
ax.scatter(x, y, color='black')
fig
```

```python
#Correspondencia- coorelacion
#Entre mas cerca se encuentren los numeros la coorelacion es mayor en porcentaje
np.corrcoef(x, y)
```

```python
# Add legend: correlation coefficient
#Asignamos a una variable la informacion redondeada
corr_ = np.round(np.corrcoef(x, y)[0][1], 2)
corr_
```

```python
#Agregar la correlacion
ax.scatter(x, y, s=300, marker='o', color='black', alpha=0.5, label=f'Correlation: {corr_}')
ax.legend(fancybox=True, prop={'size': 14}, loc='best')
ax.scatter(x, y, color='w')
fig
```
