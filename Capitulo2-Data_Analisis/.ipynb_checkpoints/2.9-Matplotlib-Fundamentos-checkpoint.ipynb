```python
#Importacion de librerias
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## Basic Plot:

```python

X = np.linspace(-np.pi, np.pi, 300, endpoint=True)#inicia en -3.14 termina en 3.14 intervalo de 300
s,c = np.sin(X), np.cos(X)# Seno y coseno en s,c
plt.figure(figsize=(8,5))#Tamaño del plot o grafica
plt.plot(X,s,color="red", linewidth=2, linestyle="--")#Formato de las lineas Y MUESTRA LA LINEA
plt.plot(X,c,color="blue", linewidth=1, linestyle="-")

plt.show()#Muestra la grafica
```

### Setting Limits:

```python
#Pone un limite a tamaño de la grafica
plt.xlim(X.min()*1.2, X.max()*1.2)
plt.ylim(c.min()*1.1, c.max()*1.1)
```

### Setting Ticks:

```python
#La informacion que se muestra  sobre el eje de las x y las y
plt.xticks( [-np.pi, -np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2, np.pi])
plt.yticks([-1, -0.5, 0, 0.5, +1])
```

### Setting tick labels:

```python
#texto enriquecido con mark-down sobre eje de las X
plt.xticks([-np.pi, -np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$-\pi/4$',r'$0$', r'$-\pi/4$',r'$+\pi/2$', r'$+\pi$'])
```

### Adding a Legend:

```python
#Agregar una legenda un "label" descriptivo
#loc='best' es el lugar de la grafica en el que se mostrara el label
plt.plot(X, c, color="blue", linewidth=2, linestyle="--", label="Cosine")
plt.plot(X, s, color="red",  linewidth=1.5, linestyle="-", label="Sine")

plt.legend(loc='best',frameon=True)
```

### Drawing Accumulative Areas:

```python
#Areas por debajo de la curva
plt.fill_between(X, 0, c, c > 0, color='blue', alpha=.25)
plt.fill_between(X, -1, s, s < 0, color='red',  alpha=.25)
```

## SubPlots and Axes:

```python
#Graficos insertados sobre un grafico mayor
#Mas de un grafico
x1 = np.linspace(0,5,20)
y1 = x1** - x1
y2 = x1 - np.sqrt(x1)
plt.subplot(1,2,1)
plt.plot(x1, y1, 'r--')
plt.subplot(1,2,2)
plt.plot(x1, y1, 'g*-')
```

### Object Oriented Method:

```python
#Metodo irientado a objetos ()
x = np.linspace(-5,5,20)
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(10, 5))#Generador de subplots

#Grafico 1
axes[0].plot(x, x**2 - x, x, x**2)
axes[0].set_title("Polynomial Functions")
#Grafico 2
axes[1].plot(x, x**2 - 0.5, x, x**3-x**2-0.5)
axes[1].set_title("Polynomial Functions")
#Grafico 3
axes[2].plot(x, x*2+0.5, x, x**2)
axes[2].set_ylim([0, 30])
axes[2].set_xlim([0, 5])
axes[2].set_title("Polynomial Functions")
```

### Text Annotation:

```python
#Texto enriquecido con markdown
axes[2].text(3, 5, r"$y=2x+0.5$", fontsize=15, color="blue")
axes[2].text(2.5, 10, r"$y=x^2$", fontsize=15, color="green")
fig#Mostrar con esta variable
```

### Saving Plots to File:

```python
#Guardar un grafico 
fig.savefig('../data/savefig.png',dpi=600)
```
