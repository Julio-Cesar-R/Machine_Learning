```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
%matplotlib inline
```

## Load the data

```python
iris = load_iris()
X=iris.data
y=iris.target
```

## Preprocessing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=2019)
```

```python
sc = StandardScaler()

X_std = sc.fit_transform(X)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
clf_lr = LogisticRegression(penalty='l2', C=1000, solver='lbfgs', multi_class='auto')
clf_dt = DecisionTreeClassifier(max_depth=3, criterion='entropy')
clf_knn = KNeighborsClassifier(n_neighbors=3)
```

## Cross Validation:

```python
clf_labels = ['Logistic Regression','Decision Tree', 'KNN']
```

```python
for clf, label in zip([clf_lr, clf_dt, clf_knn], clf_labels):
    scores = cross_val_score(estimator=clf, X=X_train_std, y=y_train, cv=10, scoring='accuracy')
    print(f"Accuracy: {scores.mean():0.2f} +/- ({scores.std():0.2f}) {label}")
```

## Majority Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe_lr = Pipeline([['sc', StandardScaler()],['clf_lr', clf_lr]])
pipe_knn = Pipeline([['sc', StandardScaler()],['clf_knn', clf_knn]])
```

```python
mv = VotingClassifier(estimators=[('lr', pipe_lr), ('dt', clf_dt), ('knn', pipe_knn)], voting='hard')
```

```python
mv = mv.fit(X_train, y_train)
```

## Cross Validation:

```python
mv_score = cross_val_score(estimator=mv,X=X_train,y=y_train,cv=10,scoring='accuracy')
```

```python
print(f"Accuracy: {mv_score.mean():0.2f} +/- ({mv_score.std():0.2f}) {'Mayority Voting'}")
```

## Drawing Classifiers

```python
res = 0.02
cl_label = ['Logistic Regression','Decision Tree','KNN','Majority Voting']
models = [clf_lr, clf_dt, clf_knn, mv]
x1min, x1max = X_train_std[:,0].min() -1, X_train_std[:,0].max() + 1
x2min, x2max = X_train_std[:,1].min() -1, X_train_std[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x1min,x1max,res),np.arange(x2min,x2max,res))
X2 = X_std[:,[2,3]]

for clf, idx, title in zip(models, np.arange(len(models)), cl_label):        
    fig = plt.figure(1,figsize=(10, 10))
    plt.subplot(2, 2, idx + 1)
    clf.fit(X2, y)
    output = clf.predict(np.array([xx.ravel(), yy.ravel()]).T)
    output = output.reshape(xx.shape)
    plt.pcolormesh(xx,yy, output, cmap=plt.cm.ocean_r)
    plt.scatter(X2[y==0, 0], X2[y==0, 1], s=30, color='blue', marker='^')
    plt.scatter(X2[y==1, 0], X2[y==1, 1], s=30, color='red', marker='o')
    plt.scatter(X2[y==2, 0], X2[y==2, 1], s=30, color='yellow', marker='x')
    plt.xlabel(iris.feature_names[2])
    plt.ylabel(iris.feature_names[3])
    plt.title(title)
              
plt.suptitle("Plotting different models")
```
