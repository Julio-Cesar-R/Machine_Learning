```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
from sklearn.manifold import TSNE
from sklearn.ensemble import RandomForestClassifier
from matplotlib.colors import ListedColormap
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Load the data

```python
iris = load_iris()
X=iris.data
y=iris.target
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df_iris = pd.DataFrame(X, columns=features)
df_iris['class'] = y
df_iris.head()
```

## Preprocessing the data

```python
sc = StandardScaler()
X_std = sc.fit_transform(X)
```

## Defining models

```python
%%time
tsne = TSNE(n_components=2, perplexity=30.0, n_iter=5000, random_state=2019)
# perplexity [5-50]-valores pequeños para grupos más unidos
X_tsne = tsne.fit_transform(X_std)
```

## Attributes:

```python
print(tsne.embedding_)
```

```python
len(tsne.embedding_)
```

```python
tsne.kl_divergence_
```

## Training the model

```python
X_train, X_test, y_train, y_test = train_test_split(X_tsne, y, test_size = 0.30, random_state = 2019)
rf = RandomForestClassifier(n_estimators=1000).fit(X_train,y_train)
```

## Metrics:

```python
pred = rf.predict(X_test)
```

```python
print(f"Accuracy:{accuracy_score(y_test,pred)}")
```

```python
print(f"Confusion Matrix:\n{confusion_matrix(y_test,pred)}")
```

```python
print(f"Classification Report:\n{classification_report(y_test,pred)}")
```

## Plotting Data:

```python
sns.set()
sns.set(rc={"figure.figsize": (10, 8)})
```

```python
PALETTE = sns.color_palette('husl', n_colors=3)
CMAP = ListedColormap(PALETTE.as_hex())
```

```python
def plot_iris_2d(x, y, title, xlabel="Dimension 1", ylabel="Dimension 2"):
    sns.set_style("whitegrid")
    
    plt.scatter(x, y,
        c=df_iris['class'],
        cmap=CMAP,
        s=70)
    
    plt.title(title, fontsize=20, y=1.05)

    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
```

```python
def plot_iris_3d(x, y, z, title):
    sns.set_style('whitegrid')
    
    fig = plt.figure(1, figsize=(8, 6))
    ax = Axes3D(fig, elev=-120, azim=120)
    
    ax.scatter(x, y, z,
               c=df_iris['class'],
               cmap=CMAP,
               s=40)
    
    ax.set_title(title, fontsize=20, y=1.05)
    
    ax.set_xlabel("Dimension 1", fontsize=16)
    ax.set_ylabel("Dimension 2", fontsize=16)
    ax.set_zlabel("Dimension 3", fontsize=16)
    
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
```

```python
plot_iris_2d(
    x = df_iris['sepal_length'],
    y = df_iris['sepal_width'],
    title = 'Plotting first two components',
    xlabel = 'Sepal length',
    ylabel = 'Sepal width')
```

```python
plot_iris_2d(
    x = X_tsne[:, 0],
    y = X_tsne[:, 1],
    title = 'Iris dataset visualized with t-SNE')
```

## Creating 3d t-SNE

```python
tsne_3 = TSNE(n_components=3, perplexity=10, n_iter=5000, random_state=2019)
X_tsne_3d = tsne_3.fit_transform(X_std)
```

## Attributes

```python
tsne_3.embedding_.shape
```

```python
tsne_3.kl_divergence_
```

## Plotting in 3D

```python
plot_iris_3d(
    x = X_tsne_3d[:,0],
    y = X_tsne_3d[:,1],
    z = X_tsne_3d[:,2],
    title = "Iris dataset visualized with t-SNE")
```

## Training the model

```python
X_train_, X_test_, y_train_, y_test_ = train_test_split(X_tsne_3d, y, test_size = 0.30, random_state = 2019)
rf_ = RandomForestClassifier(n_estimators=1000).fit(X_train_,y_train_)
```

## Metrics

```python
pred_ = rf_.predict(X_test_)
```

```python
print(f"Accuracy:{accuracy_score(y_test_,pred_)}")
```

```python
print(f"Confusion Matrix:\n{confusion_matrix(y_test_,pred_)}")
```

```python
print(f"Classification Report:\n{classification_report(y_test_,pred_)}")
```
